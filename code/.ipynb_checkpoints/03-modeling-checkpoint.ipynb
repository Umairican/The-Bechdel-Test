# We are going to attempt to create a model that can reliably predict if a script passes The Bechdel Test.

The Bechdel Test:

A film passes The Bechdel Test if it has:

1. Two (named) female characters who
2. Talk to each other about
3. Something other than a man

Seems simple, right? It's weird how few films pass this test. Let's see what the machines think of all these gendered cinema issues.
