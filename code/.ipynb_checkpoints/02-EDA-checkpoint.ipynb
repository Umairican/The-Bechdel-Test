# EDA Notebook

I decided to have a look at some of the engineered features to see what they tell us about our screenplays and what we can learn from this.

```python
import warnings #I receive an unsightly deprecation warning for matplotlib. This suppresses it
warnings.filterwarnings("ignore")

# Let's bring our mates
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
bechdel_script_df = pd.read_csv('../data/bechdel_script_df.csv')
```

```python
bechdel_script_df.head()
```

```python
bechdel_script_df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
bechdel_script_df.head()
```

```python
bechdel_script_df['simple_bechdel'].value_counts(normalize=True)
```

```python
bechdel_script_df.query('women>30')
```

It's interesting to see the films with more than 30 women named appear to be less likely to pass the Bechdel Test. Catch Me If You Can appears to have 41 named women in it. Upon further inspection, the script itself appears messy and tainted. I may have to remove it from the dataset.

```python
bechdel_script_df.query('men>60')
```

It's interesting to see the amount of named male characters in these films, as they seem to be films with a lot of locations and minor characters.

```python
plt.figure(figsize=(15,10))
sns.set_theme(style="darkgrid")
plt.xlabel('Women')
plt.ylabel('Count')
sns.histplot(data=bechdel_script_df, x='women', hue='simple_bechdel', multiple='dodge')
plt.savefig('../assets/women_count_histogram.png');
```

```python
plt.figure(figsize=(15,10))
sns.set_theme(style="darkgrid")
plt.xlabel('Men')
plt.ylabel('Count')
sns.histplot(data=bechdel_script_df, x='men', hue='simple_bechdel', multiple='dodge')
plt.savefig('../assets/men_count_histogram.png');
```

```python
plt.figure(figsize=(15,10))
plt.title('Representation of Named Men and Women in Film')
plt.xlabel('Count of Men')
plt.ylabel('Count of Women')
plt.scatter(bechdel_script_df['men'], bechdel_script_df['women'], c=bechdel_script_df['simple_bechdel'], cmap='viridis')
plt.savefig('../assets/men_women_ratio_scatter.png');
```

```python
sns.heatmap(bechdel_script_df.corr()[['simple_bechdel']].sort_values('simple_bechdel', ascending=False), annot=True,
             cmap='crest', vmax=1);
```
