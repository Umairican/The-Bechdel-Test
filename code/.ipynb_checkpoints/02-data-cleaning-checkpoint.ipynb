### RegEx and Cleaning

We are going to need to bring in some Regex to clean this all up. We have a number of line breaks and other superfluous material that make our corpus much less usable. Getting rid of these will help our model to focus on what is important.

```python
# This is code edited from my project 3 on reddit NLP
def clean_this_data_please(df):
    
    #removing &amp
    df.replace('&amp;', '', regex=True, inplace=True)
    
    #removing HTML gutter
    df.replace(to_replace=['<b>', '</b>', '<!--', '.href', '!='],   value= '', regex=True, inplace=True)
    
    #clean_df.replace('/', ' ', regex=True, inplace=True)
    
    # remove new line characters
    df.replace(to_replace=[r'\\t|\\n|\\r', '\t|\n|\r', 'if (window top)top.location=location// -->'], value='', regex=True, inplace=True)
    
    #removing other items from titles
    
    
    return df
```
